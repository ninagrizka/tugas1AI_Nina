Implementasi Algoritma Logistic Regression untuk Training Model Machine Learning

```python
# fungsi sigmoid: untuk menghitung nilai probabilitas dari target atau dependant variable (nilainya dalam rentang 0 hingga 1)

def sigmoid(x):
    
    return 1 / (1 + np.exp(-x))
```

```python
# fungsi untuk training model dengan menggunakan parameter training data, learning rate, dan jumlah iterasi

def fit(X, y, lr, iters):
    rows, cols = X.shape  # mendapatkan jumlah sampel/baris dan features/kolom dari data
    
    # inisialisasi parameter w dan b
    w = np.zeros(cols)
    b = 0
    
    # gradient descent
    for i in range(iters):
        # aproksimasi nilai y dengan menggunakan kombinasi linear dari w (weights) dan X ditambah dengan b (bias)
        linear_model = np.dot(X, w) + b
            
        # kemudian bentuk linear tadi digunakan sebagai input fungsi sigmoid
        y_predicted = sigmoid(linear_model)
        
        # hitung gradien (dw dan db)
        dw = (1 / rows) * np.dot(X.T, (y_predicted - y))
        db = (1 / rows) * np.sum(y_predicted - y)
        
        # memperbarui nilai w dan b untuk tiap iterasi
        w = w - lr * dw
        b = b - lr * db
        
    return w, b
```

```python
# fungsi untuk evaluasi/testing model sekaligus digunakan untuk prediksi
# parameternya adalah data X dan nilai w dan b yang didapat dari memanggil fungsi fit

def predict(X, w, b):
    # masih menggunakan fungsi sigmoid yang sama
    linear_model = np.dot(X, w) + b
    y_predicted = sigmoid(linear_model)
    
    # melakukan klasifikasi data output dari fungsi sigmoid yang masih berupa nilai probabilitas
    # nilai treshold default = 0.5, artinya:
    # jika nilainya kurang dari atau sama dengan 0.5 maka kategorinya adalah 0 atau 'neutral or dissatisfied'
    # jika nilainya lebih dari 0.5 maka kategorinya adalah 1 atau 'satisfied'
    y_predicted_class = [1 if i > 0.5 else 0 for i in y_predicted]
    
    return np.array(y_predicted_class)
```

```python
# fungsi untuk mengitung nilai akurasi model
# parameternya adalah nilai y sebenarnya dan y hasil prediksi model

def accuracy(y_true, y_pred):
    acc = np.sum(y_true == y_pred) / len(y_true)
    
    return acc
```

```python
# assign nilai untuk parameter learning rate dan jumlah iterasi

lr = 0.001
iters = 10000
```

```python
# memanggil fungsi untuk training model untuk mendapatkan nilai w dan b
w, b = fit(X_train, y_train, lr, iters)

# selanjutnya, nilai w dan b yang telah didapat digunakan kembali sebagai parameter untuk testing model
# dengan memanggil fungsi predict
y_predict = predict(X_test, w, b)

# mendapatkan nilai akurasi model yang telah dibuat
acc = accuracy(y_test, y_predict)
```
